```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

def plot_expert_vs_non_expert_percentage_fixed_bins(column_name, num_bins=20):
    """
    Plots the values of a specified column for experts vs. non-experts as percentages
    without a logarithmic scale on the y-axis, using the same bins for both histograms.

    Parameters:
    column_name (str): The name of the column to plot.
    num_bins (int): The number of bins to use for the histogram.
    """

    # Filter data for experts and non-experts
    experts = data[data['is.expert'] == "expert"][column_name].dropna()
    non_experts = data[data['is.expert'] == "non-expert"][column_name].dropna()

    # Determine bins using both datasets
    bins = np.histogram(np.hstack((experts, non_experts)), bins=num_bins)[1]

    # Calculate the number of observations in each group
    num_experts = len(experts)
    num_non_experts = len(non_experts)

    # Basic plot (you can customize this part)
    plt.figure(figsize=(10, 6))

    # Plot histograms and get values
    n_experts, _, _ = plt.hist(experts, alpha=0.5, label='Experts', density=False, bins=bins)
    n_non_experts, _, _ = plt.hist(non_experts, alpha=0.5, label='Non-Experts', density=False, bins=bins)

    # Normalize by dividing by the total number of observations in each group
    n_experts = n_experts / num_experts
    n_non_experts = n_non_experts / num_non_experts

    plt.xlabel(column_name)
    plt.ylabel('Percentage')
    plt.title(f'{column_name} Distribution for Experts vs. Non-Experts (Percentage)')
    plt.legend(loc='upper right')
    plt.show()

    # Print sum of percentages to verify normalization
    print(f"Sum of experts percentages: {np.sum(n_experts)}")
    print(f"Sum of non-experts percentages: {np.sum(n_non_experts)}")

# Example usage:
plot_expert_vs_non_expert_percentage_fixed_bins('Average_duration_of_fixations', num_bins=30)
```

```python
import pandas as pd

def normalize_and_compare(df, column_to_extract):
    """
    Extracts data from a specified column, groups by 'is.expert', and normalizes the results.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_to_extract (str): The name of the column to extract and analyze.

    Returns:
        dict: A dictionary containing normalized results for 'expert' and 'non-expert' groups.
              If the column is numerical, returns means. If categorical, returns value counts as percentages.
              Returns None if the specified column does not exist.
    """

    if column_to_extract not in df.columns:
        print(f"Error: Column '{column_to_extract}' not found in DataFrame.")
        return None

    grouped = df.groupby('is.expert')[column_to_extract]

    results = {}

    if pd.api.types.is_numeric_dtype(df[column_to_extract]):
        # Numerical column: calculate means
        results['expert'] = grouped.mean()['expert']
        results['non-expert'] = grouped.mean()['non-expert']

    else:
        # Categorical column: calculate percentages
        expert_counts = grouped.value_counts(normalize=True)['expert'] * 100
        non_expert_counts = grouped.value_counts(normalize=True)['non-expert'] * 100

        results['expert'] = expert_counts.to_dict()
        results['non-expert'] = non_expert_counts.to_dict()

    return results


# Example usage for numerical data:
numerical_results = normalize_and_compare(data, 'Minimum_duration_of_fixations')
if numerical_results:
    print("Numerical Data Results:")
    print(numerical_results)

# Example usage for categorical data:
categorical_results = normalize_and_compare(data, 'situation_actuelle')
if categorical_results:
    print("\nCategorical Data Results:")
    print(categorical_results)

#Example of a column that doesnt exist.
nonexistent_column = normalize_and_compare(data,'nonexistent_column')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_normalized_distribution_correct(df, column_name, bins=10):
    """
    Plots the normalized distribution of a numerical column for 'expert' and 'non-expert' groups,
    correctly displaying percentages that sum to 100%.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the numerical column to plot.
        bins (int, optional): The number of bins for the histogram. Defaults to 10.
    """

    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in DataFrame.")
        return

    if not pd.api.types.is_numeric_dtype(df[column_name]):
        print(f"Error: Column '{column_name}' is not numerical. Please provide a numerical column.")
        return

    experts = df[df['is.expert'] == 'expert'][column_name].dropna()
    non_experts = df[df['is.expert'] == 'non-expert'][column_name].dropna()

    plt.figure(figsize=(10, 6))

    n_experts, bin_edges, _ = plt.hist(experts, alpha=0.5, label='Experts', density=False, bins=bins)
    n_non_experts, _, _ = plt.hist(non_experts, alpha=0.5, label='Non-Experts', density=False, bins=bin_edges)  # Use same bins

    # Correct normalization
    n_experts_normalized = n_experts / n_experts.sum() * 100
    n_non_experts_normalized = n_non_experts / n_non_experts.sum() * 100

    # Replace original heights with normalized percentages
    plt.clf()  # Clear the previous plot

    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    plt.bar(bin_centers, n_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Experts', align='center')
    plt.bar(bin_centers, n_non_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Non-Experts', align='center')

    plt.xlabel(column_name)
    plt.ylabel('Percentage')
    # plt.xscale('log')  # Set y-axis to logarithmic scale
    plt.title(f'{column_name} Distribution for Experts vs. Non-Experts (Percentage)')
    plt.legend(loc='upper right')
    plt.show()

    # Print sum of percentages to verify normalization
    print(f"Sum of experts percentages: {np.sum(n_experts_normalized)}")
    print(f"Sum of non-experts percentages: {np.sum(n_non_experts_normalized)}")

# Load the data
data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

# Example usage for numerical data:
plot_normalized_distribution_correct(data, 'Total_duration_of_fixations', bins=50)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def plot_inset_axes(df, column_name, bins=10, inset_y_max=5):
    """
    Plots the normalized distribution with an inset axes to magnify small percentages.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the numerical column to plot.
        bins (int, optional): The number of bins for the histogram. Defaults to 10.
        inset_y_max (float, optional): The maximum y-value for the inset axes. Defaults to 5.
    """

    if column_name not in df.columns or not pd.api.types.is_numeric_dtype(df[column_name]):
        print(f"Error: Invalid column '{column_name}'.")
        return

    experts = df[df['is.expert'] == 'expert'][column_name].dropna()
    non_experts = df[df['is.expert'] == 'non-expert'][column_name].dropna()

    n_experts, bin_edges, _ = plt.hist(experts, alpha=0.5, label='Experts', density=False, bins=bins)
    n_non_experts, _, _ = plt.hist(non_experts, alpha=0.5, label='Non-Experts', density=False, bins=bin_edges)

    n_experts_normalized = n_experts / n_experts.sum() * 100
    n_non_experts_normalized = n_non_experts / n_non_experts.sum() * 100

    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.bar(bin_centers, n_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Experts', align='center')
    ax.bar(bin_centers, n_non_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Non-Experts', align='center')

    ax.set_xlabel(column_name)
    ax.set_ylabel('Percentage')
    ax.set_title(f'{column_name} Distribution (Inset Magnification)')
    ax.legend(loc='lower right')

    # Inset axes for magnification
    axins = inset_axes(ax, width="40%", height="30%", loc='upper right')
    axins.bar(bin_centers, n_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Experts', align='center')
    axins.bar(bin_centers, n_non_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Non-Experts', align='center')
    axins.set_ylim(0, inset_y_max)

    plt.show()

# Load the data
data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

# Example usage for numerical data:
plot_inset_axes(data, 'Number_of_fixations', bins=10, inset_y_max=0.02)
```

```python
def filter_dataframe_by_threshold(df, column_name, threshold):
    """
    Excludes rows from a DataFrame where a specified column exceeds a given threshold.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to filter.
        threshold (float or int): The threshold value.

    Returns:
        pd.DataFrame: A new DataFrame with rows exceeding the threshold removed.
    """

    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in DataFrame.")
        return None

    # Create a boolean mask: True for values below the threshold, False otherwise
    mask = df[column_name] <= threshold

    # Apply the mask to the DataFrame
    filtered_df = df[mask].copy() #using .copy() prevents modifying the original dataframe when working with the filtered one.

    return filtered_df

threshold = 1000
filtered_data = filter_dataframe_by_threshold(data, 'Number_of_fixations', threshold)

plot_inset_axes(filtered_data, 'Number_of_fixations', bins=10, inset_y_max=0.05)
```

```python
import pandas as pd
import numpy as np

def remove_outliers_iqr(df, column_name):
    """
    Removes outliers from a DataFrame column using the IQR method.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to remove outliers from.

    Returns:
        pd.DataFrame: A new DataFrame with outliers removed.
    """

    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in DataFrame.")
        return None

    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    filtered_df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)].copy()

    return filtered_df

# Example usage:
filtered_data_iqr = remove_outliers_iqr(data, 'Number_of_fixations')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Load the data
data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

def visualize_and_test(df, column_name):
    """
    Visualizes the distribution of a column for experts vs. non-experts using histograms
    and performs an independent samples t-test.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to visualize and test.
    """
    # Separate data for experts and non-experts
    experts = df[df['is.expert'] == 'expert'][column_name].dropna()
    non_experts = df[df['is.expert'] == 'non-expert'][column_name].dropna()

    # Plotting histograms
    plt.figure(figsize=(10, 6))

    # Use density=True to normalize the histograms
    plt.hist(experts, alpha=0.6, label='Experts', bins=30, density=True)
    plt.hist(non_experts, alpha=0.6, label='Non-Experts', bins=30, density=True)

    plt.xlabel(column_name, fontsize=12)
    plt.ylabel('Density', fontsize=12) # Update the Y-axis label
    plt.title(f'{column_name} Distribution for Experts vs. Non-Experts', fontsize=14)
    plt.legend(loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Perform t-test (this part remains the same)
    if len(experts) > 1 and len(non_experts) > 1:
        t_statistic, p_value = ttest_ind(experts, non_experts, equal_var=False) # Welch's t-test
        print(f'Results for {column_name}:')
        print(f'  T-statistic: {t_statistic:.3f}, P-value: {p_value:.3e}\n')
    else:
        print(f'Could not perform t-test for {column_name} due to insufficient data.\n')


# Example usage:
# numerical_columns = data.select_dtypes(include=['number']).columns
# for column in numerical_columns:
#     visualize_and_test(data, column)

visualize_and_test(data, 'Time_to_first_fixation')
```

```python
import pandas as pd
from scipy.stats import ttest_ind
import numpy as np # Import numpy for NaN handling

# Load the data
# Make sure the file 'path' is in the same directory
try:
    data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)
except FileNotFoundError:
    print("Error: 'path' not found.")

def evaluate_feature(df, column_name):
    """
    Evaluates a feature by performing an independent samples t-test,
    comparing group means, and returning the results as a dictionary.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to evaluate.

    Returns:
        dict: A dictionary containing the statistical results.
    """
    # Separate data for experts and non-experts
    experts = df[df['is.expert'] == 'expert'][column_name].dropna()
    non_experts = df[df['is.expert'] == 'non-expert'][column_name].dropna()

    # Check if there is enough data in both groups to perform a test
    if len(experts) < 2 or len(non_experts) < 2:
        return {
            'Column': column_name,
            'T-statistic': np.nan,
            'P-value': np.nan,
            'Mean_Experts': experts.mean() if not experts.empty else np.nan,
            'Mean_Non_Experts': non_experts.mean() if not non_experts.empty else np.nan,
            'Result': 'Not enough data for t-test'
        }

    # Perform t-test
    t_statistic, p_value = ttest_ind(experts, non_experts, equal_var=False) # Welch's t-test is often a safer default

    # Determine significance and higher group
    if p_value <= 0.05:
        if experts.mean() > non_experts.mean():
            result = "Experts significantly higher"
        else:
            result = "Non-experts significantly higher"
    else:
        result = "No significant difference"

    # Compile results into a dictionary
    results_dict = {
        'Column': column_name,
        'T-statistic': t_statistic,
        'P-value': p_value,
        'Mean_Experts': experts.mean(),
        'Mean_Non_Experts': non_experts.mean(),
        'Result': result
    }

    return results_dict

# --- Main script ---

# Get all numerical columns (excluding any ID columns if necessary)
numerical_columns = data.select_dtypes(include=['number']).columns

# Create a list to hold the results from each column evaluation
results_list = []

# Loop through each numerical column and evaluate it
for column in numerical_columns:
    # Skip columns that might not be features, like 'id'
    if 'id' in column.lower():
        continue
    results = evaluate_feature(data, column)
    results_list.append(results)

# Convert the list of dictionaries into a pandas DataFrame
results_df = pd.DataFrame(results_list)

# Set the column name as the index for better readability
results_df = results_df.set_index('Column')

# Display the final DataFrame
print("T-Test Results:")
print(results_df)

# Optionally, save the results to a CSV file
results_df.to_csv('t_test_analysis_results.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns  # For KDE

def plot_inset_axes_with_kde(df, column_name, inset_y_max=5, bin_method='freedman-diaconis'):
    """
    Plots the distribution with an inset, removes the diagonal line, and adds KDE curves.
    Corrected to use raw data for KDE.
    """

    if column_name not in df.columns or not pd.api.types.is_numeric_dtype(df[column_name]):
        print(f"Error: Invalid column '{column_name}'.")
        return

    experts = df[df['is.expert'] == 'expert'][column_name].dropna()
    non_experts = df[df['is.expert'] == 'non-expert'][column_name].dropna()

    # Automatic bin selection with limit
    if bin_method == 'freedman-diaconis':
        iqr = df[column_name].quantile(0.75) - df[column_name].quantile(0.25)
        bin_width = 2 * iqr / (len(df[column_name]) ** (1/3))
        bins = int(np.ceil((df[column_name].max() - df[column_name].min()) / bin_width))
    elif bin_method == 'sturges':
        bins = int(np.ceil(np.log2(len(df[column_name])) + 1))
    else:
        bins = 10  # Default to 10 if invalid bin_method

    bins = min(bins, 200) # Limit the maximum number of bins

    # Calculate histograms ONCE
    n_experts, bin_edges = np.histogram(experts, bins=bins)
    n_non_experts, _ = np.histogram(non_experts, bins=bin_edges)  # Use the same bins

    n_experts_normalized = n_experts / n_experts.sum() * 100
    n_non_experts_normalized = n_non_experts / n_non_experts.sum() * 100

    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.bar(bin_centers, n_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Experts', align='center')
    ax.bar(bin_centers, n_non_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Non-Experts', align='center')

    ax.set_xlabel(column_name)
    ax.set_ylabel('Percentage of Data Points per Bin')
    ax.set_title(f'{column_name} Distribution (Inset Magnification)')
    ax.legend(loc='upper right')

    # Inset axes for magnification
    axins = inset_axes(ax, width="40%", height="30%", loc='upper right')
    axins.bar(bin_centers, n_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Experts', align='center')
    axins.bar(bin_centers, n_non_experts_normalized, width=np.diff(bin_edges), alpha=0.5, label='Non-Experts', align='center')
    axins.set_ylim(0, inset_y_max)

    # Remove the diagonal line
    ax.indicate_inset_zoom(axins, edgecolor="red") #commented out, to remove the line.

    # Add KDE curves (using RAW data)
    sns.kdeplot(experts, ax=ax, color='red', label='Experts KDE')
    sns.kdeplot(non_experts, ax=ax, color='blue', label='Non-Experts KDE')

    plt.show()

# Load the data
data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

# Example usage for numerical data:
plot_inset_axes_with_kde(data, 'Time_to_first_fixation', inset_y_max=0.1, bin_method='freedman-diaconis')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns

def plot_inset_axes_with_kde(df, column_name, inset_y_max=5, bin_method='freedman-diaconis'):
    """
    Plots the distribution with an inset, removes the diagonal line, and adds KDE curves
    on a secondary y-axis.
    Corrected to use raw data for KDE and a secondary axis.
    """

    if column_name not in df.columns:
         print(f"Error: Column '{column_name}' not found in DataFrame.")
         return
    # Check if the column can be converted to numeric, handling potential errors
    try:
        # Create a copy to avoid SettingWithCopyWarning if modifying later
        numeric_col = pd.to_numeric(df[column_name], errors='coerce').copy()
    except Exception as e:
        print(f"Error converting column '{column_name}' to numeric: {e}")
        return

    if numeric_col.isnull().all():
        print(f"Error: Column '{column_name}' contains no valid numeric data after conversion.")
        return

    # Use the converted numeric column for filtering and plotting
    # Make sure 'is.expert' column exists
    if 'is.expert' not in df.columns:
        print("Error: 'is.expert' column not found in DataFrame.")
        return

    # Create temporary boolean masks for filtering
    is_expert_mask = df['is.expert'] == 'expert'
    is_non_expert_mask = df['is.expert'] == 'non-expert'

    # Filter the numeric column using the masks and drop NaNs introduced by pd.to_numeric or originally present
    experts = numeric_col[is_expert_mask].dropna()
    non_experts = numeric_col[is_non_expert_mask].dropna()


    if experts.empty and non_experts.empty:
        print(f"Error: No valid numeric data found for experts or non-experts in column '{column_name}'.")
        return
    elif experts.empty:
        print(f"Warning: No valid numeric data found for experts in column '{column_name}'.")
    elif non_experts.empty:
         print(f"Warning: No valid numeric data found for non-experts in column '{column_name}'.")


    # --- Bin Calculation (using the full valid numeric data) ---
    valid_data = numeric_col.dropna()
    if valid_data.empty:
        print(f"Error: No valid numeric data in column '{column_name}' for bin calculation.")
        return


    if bin_method == 'freedman-diaconis':
        q75 = valid_data.quantile(0.75)
        q25 = valid_data.quantile(0.25)
        iqr = q75 - q25
        n = len(valid_data)
        if iqr > 0 and n > 0:
             # Add a small epsilon to avoid division by zero if iqr is tiny
            bin_width = 2 * iqr / (n ** (1/3)) + 1e-9
            # Ensure bin_width is positive and finite
            if bin_width > 0 and np.isfinite(bin_width):
                 # Calculate range safely
                data_range = valid_data.max() - valid_data.min()
                if np.isfinite(data_range) and data_range > 0:
                    bins = int(np.ceil(data_range / bin_width))
                else:
                    bins = 10 # Fallback if range is zero or invalid
            else:
                bins = 10 # Fallback if bin_width is invalid
        else:
            bins = 10 # Fallback if IQR is 0 or no data
    elif bin_method == 'sturges':
         n = len(valid_data)
         if n > 1:
             bins = int(np.ceil(np.log2(n) + 1))
         else:
             bins = 5 # Fallback for very small data
    else:
        bins = 10  # Default

    bins = min(bins, 200) # Limit the maximum number of bins
    bins = max(bins, 5)   # Ensure a minimum number of bins

    # --- Histogram Calculation ---
    # Define bin edges based on the overall data range and calculated number of bins
    min_val, max_val = valid_data.min(), valid_data.max()
    if not np.isfinite(min_val) or not np.isfinite(max_val) or min_val == max_val:
        # If range is invalid or zero, use a default range or handle error
        print(f"Warning: Could not determine valid data range for bins in column '{column_name}'. Using defaults.")
        # Define a sensible default range if possible, or skip histogramming
        if not experts.empty:
           min_val = experts.min()
           max_val = experts.max()
        elif not non_experts.empty:
           min_val = non_experts.min()
           max_val = non_experts.max()
        else: # Should not happen if checks above passed, but as safety
             return # Cannot proceed
        if min_val == max_val: # If still single value
             min_val -= 1
             max_val +=1 # Create a small range

    # Ensure bin edges are finite
    if not np.isfinite(min_val): min_val = 0
    if not np.isfinite(max_val): max_val = min_val + bins # Default range if max is infinite

    bin_edges = np.linspace(min_val, max_val, bins + 1)


    n_experts, _ = np.histogram(experts, bins=bin_edges)
    n_non_experts, _ = np.histogram(non_experts, bins=bin_edges)

    # Normalize only if sum is non-zero
    sum_experts = n_experts.sum()
    sum_non_experts = n_non_experts.sum()

    n_experts_normalized = (n_experts / sum_experts * 100) if sum_experts > 0 else np.zeros_like(n_experts)
    n_non_experts_normalized = (n_non_experts / sum_non_experts * 100) if sum_non_experts > 0 else np.zeros_like(n_non_experts)


    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    bin_width_val = np.diff(bin_edges)[0] # Use actual calculated bin width for plotting

    # --- Plotting ---
    fig, ax = plt.subplots(figsize=(12, 7)) # Slightly larger figure

    # Plot histograms
    if sum_experts > 0:
        ax.bar(bin_centers, n_experts_normalized, width=bin_width_val, alpha=0.6, label='Experts (Hist)', align='center', color='skyblue')
    if sum_non_experts > 0:
        ax.bar(bin_centers, n_non_experts_normalized, width=bin_width_val, alpha=0.6, label='Non-Experts (Hist)', align='center', color='lightcoral')

    ax.set_xlabel(column_name, fontsize=12)
    ax.set_ylabel('Percentage of Data Points per Bin (%)', fontsize=12, color='black')
    ax.set_title(f'{column_name} Distribution (Experts vs. Non-Experts)', fontsize=14)
    ax.tick_params(axis='y', labelcolor='black')
    ax.grid(axis='y', linestyle='--', alpha=0.7)


    # Create secondary axis for KDE
    ax2 = ax.twinx()

    # Plot KDE curves on the secondary axis
    kde_color_expert = 'blue'
    kde_color_non_expert = 'red'
    if not experts.empty:
        sns.kdeplot(experts, ax=ax2, color=kde_color_expert, label='Experts (KDE)', linewidth=2)
    if not non_experts.empty:
        sns.kdeplot(non_experts, ax=ax2, color=kde_color_non_expert, label='Non-Experts (KDE)', linewidth=2)

    ax2.set_ylabel('Density (KDE)', fontsize=12, color='darkred') # Label for secondary axis
    ax2.tick_params(axis='y', labelcolor='darkred')
    ax2.grid(False) # Turn off grid for secondary axis if desired


    # Combine legends from both axes
    lines, labels = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper right')
    # Hide the automatic legend from the primary axis if combined on ax2
    # ax.get_legend().remove()


    # --- Inset Axes ---
    # Create inset axes relative to ax (the primary axes)
    axins = inset_axes(ax, width="40%", height="30%", loc='center right') # Try simpler location # Adjusted loc/borderpad

    # Plot histograms on inset (using the same data)
    if sum_experts > 0:
        axins.bar(bin_centers, n_experts_normalized, width=bin_width_val, alpha=0.6, align='center', color='skyblue')
    if sum_non_experts > 0:
        axins.bar(bin_centers, n_non_experts_normalized, width=bin_width_val, alpha=0.6, align='center', color='lightcoral')

    # Set inset limits (use dynamic range based on data if inset_y_max is too small)
    max_hist_val_inset = max(n_experts_normalized.max() if sum_experts > 0 else 0,
                             n_non_experts_normalized.max() if sum_non_experts > 0 else 0)

    # Determine reasonable x-limits for inset based on where the bulk of the data lies, e.g., 95th percentile
    x_lim_inset_max = valid_data.quantile(0.95) if not valid_data.empty else bin_centers.max()
    # Or focus on a specific range if needed, e.g., near zero if 'Time_to_first_fixation'
    x_lim_inset_max = min(x_lim_inset_max, bin_centers.max()) # Ensure it doesn't exceed overall max


    # Ensure inset_y_max is sensible compared to data
    actual_inset_y_max = min(inset_y_max, max_hist_val_inset * 1.1) if max_hist_val_inset > 0 else inset_y_max
    actual_inset_y_max = max(actual_inset_y_max, 0.01) # Ensure a minimum height

    axins.set_xlim(bin_centers.min(), bin_centers.max()) # Adjust x-limits for inset focus
    axins.set_ylim(0, actual_inset_y_max)
    axins.set_title("Inset View", fontsize=10)
    axins.tick_params(axis='both', which='major', labelsize=8)

    # Indicate zoom - connect `ax` (original) to `axins` (inset)
    # Use mark_inset which draws the box and connectors
    # Ensure you pass the parent axes (ax) and the inset axes (axins) correctly
    # Also specify the location corners (e.g., 2 and 4 are bottom-left and top-right corners of the zoom area)
    mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="red", lw=1)


    # Optional: Add a vertical line at zero if relevant (e.g., for time)
    if min_val < 0 < max_val:
       ax.axvline(0, color='grey', linestyle=':', linewidth=1)


    fig.set_constrained_layout(True) # Adjust layout to prevent overlap
    plt.show()

# --- Need to import mark_inset ---
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

# --- Load Data ---
try:
    data = pd.read_csv('path', delimiter="\t", quoting=0, low_memory=False)

    plot_inset_axes_with_kde(data, 'Time_to_first_fixation', inset_y_max=0.1) # Adjusted inset_y_max for percentage scale

except FileNotFoundError:
    print("Error: Data file 'path' not found.")
except pd.errors.EmptyDataError:
    print("Error: Data file is empty.")
except Exception as e:
    print(f"An error occurred loading or processing the data: {e}")
```
